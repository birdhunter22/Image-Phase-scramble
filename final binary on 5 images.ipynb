```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from random import randrange
```

```python
img1 = cv.imread('dog.jpg',0)
img1 = cv.resize(img1, (400, 400)) 
img2 = cv.imread('bottle.jpg',0)
img2 = cv.resize(img2, (400, 400))
img3 = cv.imread('chair.jpeg',0)
img3 = cv.resize(img3, (400, 400))
img4 = cv.imread('car.jpg',0)
img4 = cv.resize(img4, (400, 400))
img5 = cv.imread('mona.jpg',0)
img5 = cv.resize(img5, (400, 400))
```

```python
fig=plt.figure(figsize=(20, 20))
fig.add_subplot(1, 5, 1)
plt.imshow(img1,cmap="gray")
fig.add_subplot(1, 5, 2)
plt.imshow(img2,cmap="gray")
fig.add_subplot(1, 5, 3)
plt.imshow(img3,cmap="gray")
fig.add_subplot(1, 5, 4)
plt.imshow(img4,cmap="gray")
fig.add_subplot(1, 5, 5)
plt.imshow(img5, cmap="gray")


plt.show()
```

```python
print img1.shape
```

```python
f1 = np.fft.fft2(img1)
f2 = np.fft.fft2(img2)
f3 = np.fft.fft2(img3)
f4 = np.fft.fft2(img4)
f5 = np.fft.fft2(img5)
```

```python
m1=np.abs(f1)
m2=np.abs(f2)
m3=np.abs(f3)
m4=np.abs(f4)
m5=np.abs(f5)
```

```python
p1=np.angle(f1)
p2=np.angle(f2)
p3=np.angle(f3)
p4=np.angle(f4)
p5=np.angle(f5)
```

```python
r= np.ravel(p1)
for g in range(1,20):
    magnitude = np.abs(f1)
    r= np.ravel(r)
    for i in range(0,4000):
        ri1 = randrange(len(r))
        ri2 = randrange(len(r))
        #print r[ri1], r[ri2]
        t=r[ri1]
        r[ri1]=r[ri2]
        r[ri2]=t
    r = np.reshape(r, (400, 400))    
    plt.figure(figsize=(10,10))    
    #original image    
    original= np.abs(np.fft.ifft2(magnitude*np.exp(1j*(np.angle(f1)))))
    #thresholding the original image
    thresh = 127
    original = cv.threshold(original, thresh, 255, cv.THRESH_BINARY)[1]
    plt.subplot(121),plt.imshow(original, cmap = 'gray')
    plt.title('original image '), plt.xticks([]), plt.yticks([])
    #generating the image with random phase
    random_phase_img= np.abs(np.fft.ifft2(magnitude*np.exp(1j*(r))))
    q=5*g
    random_phase_img = cv.threshold(random_phase_img, thresh, 255, cv.THRESH_BINARY)[1]
    plt.subplot(122),plt.imshow(random_phase_img, cmap = 'gray')
    plt.title('  &   random phase image with '+ str(q) +"% random phase"), plt.xticks([]), plt.yticks([])
    plt.show()
```
